# LED Sources

## Flame Spectrometer

```julia
dark = "../data/led_sources/dark.txt"
uv   = "../data/led_sources/405nm_withdark.txt"
blue = "../data/led_sources/470nm_withdark.txt"
green= "../data/led_sources/533nm_withdark.txt"
red  = "../data/led_sources/635nm_withdark.txt"

function sum_gaussians(x, A, mu, sigma, c)
    return A .* exp.( .- ( (x .- mu)./sigma ).^2 ) .+ c
end
```

```julia
#| output: false

df = DataFrame(CSV.File(dark,delim='\t',skipto=15))
rename!(df, 1=>:wavelengths, 2=>:counts)
Î»s = df.wavelengths
dark_data = df.counts

df = DataFrame(CSV.File(uv,delim='\t',skipto=15))
rename!(df, 1=>:wavelengths, 2=>:counts)
uv_data = df.counts
uv_data .-= dark_data
uv_data ./= maximum(uv_data)

df = DataFrame(CSV.File(blue,delim='\t',skipto=15))
rename!(df, 1=>:wavelengths, 2=>:counts)
blue_data = df.counts
blue_data .-= dark_data
blue_data ./= maximum(blue_data)

df = DataFrame(CSV.File(green,delim='\t',skipto=15))
rename!(df, 1=>:wavelengths, 2=>:counts)
green_data = df.counts
green_data .-= dark_data
green_data ./= maximum(green_data)

df = DataFrame(CSV.File(red,delim='\t',skipto=15))
rename!(df, 1=>:wavelengths, 2=>:counts)
red_data = df.counts
red_data .-= dark_data
red_data ./= maximum(red_data)

_
```

```julia
Î»s_used = Î»s[10:end-10]
plot_obj = plot(Î»s_used,uv_data[10:end-10],color=:magenta,label="UV LED")
plot!(Î»s_used,sum_gaussians(Î»s_used,1,405,17,0),color=:magenta,linestyle=:dash,label="UV bell")
plot!(Î»s_used,blue_data[10:end-10],color=:blue,label="Blue LED")
plot!(Î»s_used,sum_gaussians(Î»s_used,1,470,18,0),color=:blue,linestyle=:dash,label="Blue bell")
plot!(Î»s_used,green_data[10:end-10],color=:green,label="Green LED")
plot!(Î»s_used,sum_gaussians(Î»s_used,1,533,36,0),color=:green,linestyle=:dash,label="Green bell")
plot!(Î»s_used,red_data[10:end-10],color=:red,label="Red LED")
plot!(Î»s_used,sum_gaussians(Î»s_used,1,635,14,0),color=:red,linestyle=:dash,label="Red bell")

xlabel!(plot_obj,"wavelength (nm)")
ylabel!(plot_obj,"normalized amplitude")
xlims!(plot_obj,(350,750))
title!(plot_obj,"Flame Spectrometer")
```

# Finding pulses

```julia
function find_rising_edges(timeseries)
    # Initialize an empty array to hold the indices of rising edges
    rising_edges = Int[]

    # Loop over the array (starting from the second element)
    for i in 2:length(timeseries)
        # Check if the current value is greater than the previous value
        if timeseries[i] > timeseries[i-1]
            # If it is, that's a rising edge, so add the index to the list
            push!(rising_edges, i)
        end
    end

    return rising_edges
end

# Example usage:
timeseries = [1, 2, 1, 3, 2, 5, 4, 6]
edges = find_rising_edges(timeseries)
println(edges)  # This will print the indices of the rising edges
```
